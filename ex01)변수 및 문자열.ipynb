```python
# 문자열 포메팅
# - 문자열 안의 특정한 값을 바꿔야 할 경우가 있을때 (변수)
```

```python
s1 = '오늘은 10월 21일 입니다.'

s1 = '오늘은 10월 22일 입니다.'

print(s1)
```

```python
# % 기호를 상용하는 방법

day = 21

# 정수형 변수를 넣을 자리에 %d , 마지막에 % (들어갈 변수면)

s1 = '오늘은 10월 %d일 입니다.'%day

# %d : 정수형
# %f : 실수형
# %s : 문자형

print(s1)
```

```python
day += 1
# day = day+1

print('오늘은 10월 %d일 입니다.'%day)
```

```python
# format 함수를 사용한 포메팅 (예전 방법)

month = 10
day = '오'

# format(첫번째 변수명, 두번째 변수명.....)
s = '오늘은 {}월 {}일 입니다.'.format(month,day)
print(s)
```

```python
# f문자열을 사용한 포메팅
# 현재 가장 많이 쓰이는 방법, python3 부터 제공하는 기능 

month = 10
day = 21

s = f'오늘은 {month}월 {day}일 입니다.'
print(s)
```
